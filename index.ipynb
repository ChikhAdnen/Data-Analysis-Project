# NBA Shooting Data
This dataset contains shooting statistics for four different players during the 2021 NBA Playoffs.

```python
import pandas as pd
import matplotlib.pyplot as plt

nba=pd.read_csv("nba_players_shooting.csv", index_col=0)
nba
```

## Data Dictionary
|variable |class     |description                                                  |
|:--------|:---------|:------------------------------------------------------------|
|SHOOTER  |String    |Name of the player taking the shot                           |
|X        |float     |Horizontal distance of the shot taken from the basket in ft  |
|Y        |float     |Vertical distance of the shot taken from the basket in ft    |
|RANGE    |String    |Radius range of the shot taken from the basket in ft         |
|DEFENDER |String    |Name of the player defending the shot                        |
|SCORE    |String    |'MADE' if shot is scored, else 'MISSED'                      |

[Source](https://www.nba.com/stats/players/shooting/?Season=2020-21&SeasonType=Playoffs&PerMode=Totals) of dataset.

```python
seth_curry=nba[nba["SHOOTER"]=="Seth Curry"]
chris_paul=nba[nba["SHOOTER"]=="Chris Paul"]
russel_westbrook=nba[nba["SHOOTER"]=="Russell Westbrook"]
trae_young=nba[nba["SHOOTER"]=="Trae Young"]


```

### Christopher Emmanuel Paul (born May 6, 1985), nicknamed "CP3" and The Point God, is an American professional basketball player who plays for the Phoenix Suns of the National Basketball Association (NBA). Paul is widely regarded as one of the greatest point guards of all time .
<img src="images/Chris-Paul.webp" alt="Chris paul" width="100"/>

```python

score=chris_paul.groupby(["SCORE"])["SCORE"].count()
sr=(score["MADE"]/(score["MADE"]+score["MISSED"]))*100
print("FG% =",sr,"%")
cp_made=chris_paul[chris_paul["SCORE"]=="MADE"]
print("Total shots made",cp_made["SCORE"].count())
cp_missed=chris_paul[chris_paul["SCORE"]=="MISSED"]
fig = plt.figure()
ax = fig.add_subplot( title="Chris Paul")
ax.scatter(cp_made.X,cp_made.Y,alpha=0.4,c="g")
ax.scatter(cp_missed.X,cp_missed.Y,alpha=0.4,c="r")
```
